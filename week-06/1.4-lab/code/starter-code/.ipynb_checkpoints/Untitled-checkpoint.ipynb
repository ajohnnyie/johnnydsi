```python
f = open('aa.txt')
x = f.readlines()
```

```python
import pandas as pd
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer('\w+|\$[\d\.]+|\S+')
tokens = [tokenizer.tokenize(i) for i in x]
tokens
```

```python
namelList=[]
emailList=[]
i=0
for i in tokens[0]:
    if "@" in i:
        emailList.append(i)
    if "-space:nowrap;\\" in i:
        namelList.append(i)
    
```

```python
a=pd.DataFrame(emailList)
```
