```python
import pandas as pd
import numpy as np


df = pd.read_csv('billboard.csv')
```

```python
df.rename(columns={'artist.inverted':'Artist Name'}, inplace=True)
df.rename(columns={'date.entered':'Date Entered'}, inplace=True)
df.rename(columns={'date.peaked':'Date Peaked'}, inplace=True)
df.rename(columns={'year':'Year'}, inplace=True)
df.rename(columns={'genre':'Genre'}, inplace=True)
df.rename(columns={'time':'Time'}, inplace=True)
df.rename(columns={'track':'Track'}, inplace=True)

import re

def removeAsciChars(x):
    return re.sub(r'[^\x00-\x7f]',r'',x)

df['Track'] = df['Track'].apply(removeAsciChars)
df['Artist Name'] = df['Artist Name'].apply(removeAsciChars)


#df['x75th.week']=df['x75th.week'].fillna("Out")
df
```

```python

df1=pd.melt(df, id_vars=['Year','Artist Name','Track','Time','Genre','Date Entered','Date Peaked'],var_name='Week', value_name='Ranking')
df2=df1.dropna(subset=['Ranking'], how='all')
df2
```

```python
#Save the dataframe to csv file

df2.to_csv('dataab')
```

```python
#Tring to figure out the days it takes for a track to get to the peak


import datetime


#def getDobFromStr(d): 
#    return datetime.datetime.strptime(d,"%Y-%m-%d").date()
#def getMonthFromStr(d): 
#    return datetime.datetime.strptime(d,"%Y-%m-%d").month

#df2['Date Entered'] = df2['Date Entered'].apply(getDobFromStr)
#df2['Date Peaked'] = df2['Date Peaked'].apply(getDobFromStr)
#df2['Date Diff'] = df2.apply( lambda x: findDeltaDate( x['Date Entered'], x['Date Peaked'] ) )




#print df2.columns

def findDeltaDate(x):    
    return (x[6] - x[5]).days

#df2.dtypes

df2['Days to Peak'] = df2.apply(findDeltaDate, axis=1)

df2
```

```python
#This shows the amount of weeks the songs stay on the billboard by a histogram
#From the plot, we can see that most sonnds stay on the billboard for about 20 weeks.

dfGrouped = df2.groupby('Track')


columnNames = []
counts = []
for key, group in dfGrouped:
    columnNames.append( str(key) )
    counts.append( len(group) )


import matplotlib.pyplot as plt

dfToPlot = pd.DataFrame( {"track_name" : columnNames, "counts": counts } )
dfToPlot.plot(kind = 'hist'); plt.axhline(0, color='k')
plt.show()

```
